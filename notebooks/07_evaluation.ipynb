```python
# ============================================
# CELL 7: Custom Job Description Creator
# ============================================

def create_custom_job_description():
    """Helper to create custom job description from user input"""
    import streamlit as st
    
    with st.expander("✏️ Or Create Custom Job Description"):
        col1, col2 = st.columns(2)
        
        with col1:
            job_title = st.text_input("Job Title", placeholder="e.g., Senior Python Developer")
            skills = st.text_area("Required Skills (comma-separated)", 
                                placeholder="Python, Django, SQL, AWS, REST APIs")
            experience = st.text_input("Experience Required", placeholder="3-5 years")
        
        with col2:
            responsibilities = st.text_area("Key Responsibilities", 
                                            placeholder="• Develop web applications\n• Write clean code\n• Collaborate with team")
            education = st.text_input("Education Required", placeholder="Bachelor's in CS or equivalent")
        
        additional = st.text_area("Additional Requirements/Qualifications")
        
        if st.button("Generate Job Description"):
            job_desc = f"""
            Job Title: {job_title}
            
            Experience Required: {experience}
            Education: {education}
            
            Required Skills: {skills}
            
            Key Responsibilities:
            {responsibilities}
            
            Additional Requirements:
            {additional}
            """
            
            st.success("✅ Custom job description created!")
            return job_title, job_desc
    
    return None, None

print("✅ Custom job creator ready!")
```
